```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np
import re

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots we will create
```

```python
# Read in the data
df = pd.read_csv(r"C:\Users\B\Documents\Portfolio_projects\movies.csv")
```

```python
# Let's look at the data

df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print ('{} - {}%'.format(col, pct_missing))
```

```python
# Data Types for our columns
df.dtypes
```

```python
# Drop na and change column data types
df = df.dropna(subset=['budget', 'gross', 'votes'])
df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
df['votes'] = df['votes'].astype('Int64')
```

```python
df.tail()
```

```python
# Create Correct Year Column

df['yearcorrect'] = df['released'].astype(str).str.extract(r'(\d{4})')
df.head()
```

```python
df['yearcorrect'] = df['yearcorrect'].astype(int)
```

```python
# Arrange in ascending order according to gross

df.sort_values(by=['gross'], inplace = False, ascending=False)
```

```python
#Drop any duplicates

df.drop_duplicates()
```

```python
# Scatter Plot with budget vs gross
plt.scatter(x=df['budget'] / 1_000_000, y=df['gross'] / 1_000_000_000)
plt.xlabel("Budget (In Millions $)")
plt.ylabel("Gross Earnings (In Billions $)")
plt.title("Budget VS Gross Earnings")
plt.show();
```

```python
# Plot budget vs gross using seaborn

plt.figure()
sns.regplot(
    x=df['budget'] / 1_000_000,
    y=df['gross'] / 1_000_000_000,
    data=df,
    scatter_kws={"alpha":0.5, "color": "red"},
    line_kws={"color": "blue"}
)
plt.xlabel("Budget (Millions)")
plt.ylabel("Gross (Billions)")
plt.title("Regression Plot: Budget vs Gross")
plt.show();
```

```python
# Correlation
df.corr(method='kendall', numeric_only=True) # pearson, kendall, spearman
```

```python
df.corr(method='spearman', numeric_only=True)
```

```python
df.corr(method='pearson', numeric_only=True)
```

```python
corr_matrix = df.corr(method='pearson', numeric_only=True)

sns.heatmap(corr_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Features")
plt.show()
```

```python
# Look at company
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
corr_matrix = df_numerized.corr(method='pearson', numeric_only=True)

sns.heatmap(corr_matrix, annot=True)

plt.title("Correlation Matrix")
plt.show()
```

```python
df_numerized.corr()
```

```python
corr_mat = df_numerized.corr()
corr_pairs = corr_mat.unstack()
sorted_pairs = corr_pairs.sort_values(kind='quicksort')
high_corr = sorted_pairs[(sorted_pairs > 0.5) & (sorted_pairs < 1)]
high_corr
```
